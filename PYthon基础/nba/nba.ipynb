```python
'''
Time:2023.8.22
nba球星信息
'''
from lxml import etree
#发送地址
url = 'https://nba.hupu.com/stats/players'

headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36 Edg/115.0.1901.203'}

#发送请求
 
#pip install requests
import requests
resp = requests.get(url,headers = headers)#resp接受响应
print( resp.text)
#处理结果
e = etree.HTML(resp.text)
# 解析响应数据
names = e.xpath('//table[@class="players_table"]//tr/td[2]/a/text()')
teams = e.xpath('//table[@class="players_table"]//tr/td[3]/a/text()')

with open('nba.txt','w',encoding='utf-8') as f:
#是否保存
    for name,team in zip(names,teams):
        f.write(f'姓名：{name} 团队：{team} \n')
```
